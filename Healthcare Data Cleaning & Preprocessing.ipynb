```python
import pandas as pd
import numpy as np
from dateutil import parser
pd.set_option('display.max_columns', None)


df = pd.read_csv('/Users/danon/Downloads/advanced_dirty_healthcare_data.csv')

#Create a copy of the data
data = df.copy()
```

```python
#Got rid of the 'P' in the patient ID code and formatted it as a numerical datatype for Patient ID

# P100000 -> 100000

data['Patient_ID'] = data['Patient_ID'].str.lstrip('P').astype(int)
data
```

```python
#Normalised the Gender column

data['Gender'].unique()

clean_gender = {'M':'Male' , 'F':'Female'}
data['Gender'] = data['Gender'].replace(clean_gender)
data
```

```python
#Check for any negative ages, NaN values or unrealistic ages

data[data['Age']<0]
data[data['Age'].isna()]
data[data['Age'] > 120]
```

```python
data['Gender'].unique()

#array(['Male', 'Female', 'Other', nan], dtype=object)

#The Gender column is already normalised.

data['Department'].unique()

#array(['Oncology', 'Pediatrics', nan, 'Emergency', 'Cardiology', 'Orthopedics'], dtype=object)

#The Department column is already normalised.

data['Diagnosis'].unique()

#array(['Diabetes', 'Cancer', 'Stroke', 'Hypertension', 'Asthma', nan], dtype=object)

#The Diagnosis column is already normalised
```

```python
#  Try to convert Visit_Date to datetime, coercing errors to NaT (Not a Time)

data['Visit_Date'] = pd.to_datetime(data['Visit_Date'], errors = 'coerce')
data['Discharge_Date'] = pd.to_datetime(data['Discharge_Date'], errors = 'coerce')


# Ensuring these rows have no Na Values

print(data['Visit_Date'].notna().sum())
print(data['Visit_Date'].isna().sum())


print(data['Discharge_Date'].notna().sum())
print(data['Discharge_Date'].isna().sum())
```

```python
data[data['Discharge_Date'] < data['Visit_Date']]

#The data reveals data inconsistencies, as Discharge Date for certain rows have dates that are earlier than their Visit Date counterpart.

#To normalise this, all Discharge Dates that occur before the Visit Dates will be turned into Discharge Dates instead. 
#Otherwise, they will remain the same



#np.where(condition, value_if_true, value_if_false)


data['Discharge_Date'] = np.where(
    data['Discharge_Date'] < data['Visit_Date'],  # Condition: Discharge is before Visit
    data['Visit_Date'],  # Action: Replace with Visit_Date
    data['Discharge_Date']  # Otherwise, keep the original discharge date
)

data
```

```python
data['Status'].unique()

#array(['Discharged', 'In Treatment', 'Admitted', 'Unknown'], dtype=object)

#The Status column is already normalised

data['Notes'].unique()

#array(['Follow-up needed', 'Needs review', nan, 'Referred to specialist'], dtype=object)

#The Notes column is already normalised
```

```python
#Blood pressure is formatted as '160/100'. We will further categorise the blood pressure into 'Systolic' and 'Distolic' by 
#splitting up the blood pressure values into separate columns called 'Systolic_Bp' and 'Distolic_Bp'.


data[['Systolic_Bp', 'Diastolic_Bp']] = data['Blood_Pressure'].str.split('/', expand=True)
data['Systolic_Bp'] = pd.to_numeric(data['Systolic_Bp'], errors='coerce')
data['Diastolic_Bp'] = pd.to_numeric(data['Diastolic_Bp'], errors='coerce')


data
```

```python
print(data['Heart_Rate'].unique())

#The heart rate column seems to be cleaned with the exception of a potential outlier occurence of a '300' heartrate.

```

```python
#Negative BMI values will be returned as NaN

data[data['BMI'] < 0]

def clean_bmi (BMI):

    return np.nan if BMI < 0 else BMI

data['BMI'] = data['BMI'].apply(clean_bmi)

data
```

```python
data['Smoker_Status'].unique()

#array(['No', 'Yes', 'Former', nan, 'Unknown'], dtype=object)

#The Smoker status column is already normalised

data['Insurance_Type'].unique()

#array([nan, 'Private', 'Public'], dtype=object)

#The Insurance Type column is already normalised

```

```python
#'W' has been normalised to 'White' ethnicity, everything else is already normalised

data['Ethnicity'].unique()

clean_ethnicity = {'W': 'White'}

data['Ethnicity'] = data['Ethnicity'].replace(clean_ethnicity)
data
```

```python
print(data['Medication_Count'].unique())


#The medication counts are also positive integers, so is normalised
```

```python
#Before, some of the discharge dates were earlier than visit dates. As a result, the length of stay may have contained negative numbers
#However, as the dates were cleaned and updated, we will now ensure the length of stay is also updated to be a positive value

data['Length_of_Stay'] = data['Discharge_Date'] - data['Visit_Date']

data
```

```python
#We will now transform the 'Length of Stay' column into a numerical data type



def clean_stay(stay):
    if pd.isna(stay):
        return np.nan

    # Keep only the first number before the space
    stay = str(stay).split()[0]

    try:
        return float(stay)
    except ValueError:
        return np.nan

data['Length_of_Stay'] = data['Length_of_Stay'].apply(clean_stay)
data
```
